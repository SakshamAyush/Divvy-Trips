```python
import pandas as pd
#import geopy
import numpy as np
#import matplotlib.pyplot as plt
import datetime
from datetime import datetime
```

```python
#Reading CSV files of Divvy Data
df1 = pd.read_csv('../../Divvy_Trips_2019_Q1.csv')
df2 = pd.read_csv('../../Divvy_Trips_2019_Q2.csv')
df3 = pd.read_csv('../../Divvy_Trips_2019_Q3.csv')
df4 = pd.read_csv('../../Divvy_Trips_2019_Q4.csv')
```

```python
df2.columns = ['trip_id', 'start_time', 'end_time', 'bikeid', 'tripduration',
       'from_station_id', 'from_station_name', 'to_station_id',
       'to_station_name', 'usertype', 'gender', 'birthyear']
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
```

```python
df2.head()
```

```python
df = pd.concat([df1,df2,df3,df4])
```

```python
df.head()
```

```python
print(df.shape)
```

```python
df.drop(columns =['bikeid','to_station_id','to_station_name','end_time'], axis=1,inplace=True)
```

```python
df['start_time'] = df['start_time'].astype('datetime64[ns]')
```
